```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
url = "https://en.wikipedia.org/wiki/Apollo_program"
page = urlopen(url)
page = page.read().decode("utf-8")
```

```python
soup = BeautifulSoup(page, 'html.parser')
```

```python
text = soup.get_text()
```

```python
paras = ["".join(list(s.strings)).strip()+"\n" for s in soup.find_all('p')]
paras = [s for s in paras if s != "\n"] # paras
```

```python
paras
```

```python
cleaned_text = "\n".join(paras)
print(cleaned_text)

with open("clean_page.txt", "w") as f:
    f.write(cleaned_text)
```

```python
words = [w for w in word_tokenize(cleaned_text) if w.isalnum()] # word tokenize
sentences = sent_tokenize(cleaned_text) # sentence tokenize
```

```python
unique_words = set(words) # unique words
```

```python
links = [link.get('href') for link in soup.find_all('a')]
links = [link for link in links if link is not None and link[:4] == "http"]
```
